# (iv)Aggregation & Grouping

```python
dataset.groupby("Status")["Amount"].agg(["mean", "sum", "count"])
#group by and calculate mean,sum,count
```

```python
dataset.groupby(["Status", "Amount"]).agg({"Amount": "sum", "Amount": "mean"})
#multiple group aggregations
```

```python
dataset.pivot_table(values="Amount", index="Status", columns="Category", aggfunc="sum") #pivot table
# values='Score': The data you want to aggregate
# index='Name': Rows will be grouped by Name
# columns='Subject': Columns will be grouped by Subject
# aggfunc='mean': Aggregation function (can be 'sum', 'count', 'max', etc.)

# Customizations
# fill_value=0: Replace missing values with 0
# margins=True: Add totals row/column
# dropna=False: Keep rows with missing data
```

# (v) Merging & Joining

```python
df1=dataset["Category"]
df2=dataset["Status"]
```

```python
pd.concat([df1, df2])
# stack dataframes vertically or horizontally
```

```python
df1=dataset[["Category","Date","Fulfilment","Order ID"]]
df2=dataset[["Order ID","Status"]]
```

```python
pd.concat([df1, df2])
# stack dataframes vertically or horizontally
#note both order id add on 
```

```python
pd.merge(df1, df2, on="Order ID", how="inner")
# SQL-style joins (inner, left, right, outer)
```

# row filtering

```python
dataset_filtered = dataset[dataset['Date'] > "2022-05-03"]
```

```python
dataset_filtered
```

```python
# Age > 30 AND Salary > 50000
dataset_filtered = dataset[(dataset['Date'] > "2022-05-03") & (dataset['Status'] == "Shipped")]
```

```python
dataset_filtered
```

```python
# Age > 30 OR Salary > 50000
dataset_filtered = dataset[(dataset['Date'] > "2022-05-03") | (dataset['Status'] == "Shipped")]
```

```python
dataset_filtered
```

```python
# Filter using query() (more readable)
dataset_filtere = dataset.query("Date > '2022-05-03' and Amount > 400")
```
